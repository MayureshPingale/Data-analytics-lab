```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_train = pd.read_csv("Train.csv")
df_train.head()
```

```python
df_train.isnull().sum()
```

```python
print(df_train['Item_Weight'].unique())
print(df_train['Outlet_Size'].unique())
print(df_train.shape)
```

```python
df_train.groupby(by="Outlet_Location_Type").Item_Outlet_Sales.sum().head()
```

```python
df_train.groupby(by="Outlet_Type").Item_Outlet_Sales.sum().head()
```

```python
df_train['Outlet_Size'].fillna("Not Available",inplace=True)
df_train.isnull().sum()
```

```python
df_train['Item_Weight'].fillna(df_train['Item_Weight'].mean(),inplace=True)
df_train.isnull().sum()
```

```python
df_train.drop("Item_Identifier",axis=1,inplace=True)
df_train.drop("Outlet_Identifier",axis=1,inplace=True)
df_train.drop("Item_Fat_Content",axis=1,inplace=True)
df_train.drop("Item_Type",axis=1,inplace=True)
df_train.drop("Item_Visibility",axis=1,inplace=True)
df_train.head()
```

```python
df_train = pd.get_dummies(df_train, prefix=['Outlet_Size', 'Outlet_Location_Type','Outlet_Type'])
df_train.head()
```

```python
df_train.drop("Outlet_Size_Not Available",axis=1,inplace=True)
df_train.drop("Outlet_Location_Type_Tier 3",axis=1,inplace=True)
df_train.drop("Outlet_Type_Supermarket Type3",axis=1,inplace=True)
df_train.head()
```

```python
y = df_train['Item_Outlet_Sales'].values
df_train.drop("Item_Outlet_Sales",axis=1,inplace=True)
X = df_train.iloc[:,:].values 
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size = 0.2)
```

```python
from sklearn.linear_model import LinearRegression
linear = LinearRegression()
linear.fit(X_train,y_train)
y_pred = linear.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
print("MAE",mean_absolute_error(y_true=y_test,y_pred=y_pred))
```

```python
from sklearn.svm import SVR
Svr = SVR(kernel='rbf')
Svr.fit(X_train,y_train)
y_pred =Svr.predict(X_test)
from sklearn.metrics import mean_absolute_error
print("MAE",mean_absolute_error(y_true=y_test,y_pred=y_pred))
```
